```python
#import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#creat dataframes using team and player csv's
csv_file_1 = "./resources/worldcup_teams.csv"
teams_df = pd.read_csv(csv_file_1)

csv_file_2 = "./resources/worldcup_players.csv"
players_df = pd.read_csv(csv_file_2)
```

```python
list(teams_df.columns.values)
```

```python
#clean team data, delete undesired columns
updated_teams_df = teams_df[[
       'team_name', 'season', 'country', 'matches_played',
       'wins', 'draws', 'losses', 'points_per_game',
       'league_position', 'performance_rank', 'goals_scored', 'goals_conceded',
       'goal_difference', 'total_goal_count', 'clean_sheets', 'corners_total',
       'cards_total', 'average_possession', 'shots', 'fouls', 'average_total_goals_per_match']].copy()
updated_teams_df.head()
```

```python
#clean players data, delete undesired columns
updated_players_df = players_df[[
       'full_name','age', 'league', 'season', 'position', 'Current Club',
       'minutes_played_overall', 'nationality', 'appearances_overall', 
       'goals_overall', 'assists_overall', 'penalty_goals',
       'penalty_misses', 'clean_sheets_overall', 'conceded_overall', 
       'yellow_cards_overall', 'red_cards_overall']].copy()
updated_players_df.head()
```
